```python
import pandas as pd
import os
```

```python
# read label
old_data_dir = '../data/ft_info/'
tar_path = os.path.join(old_data_dir, "full-text_sub_set_500_a_2nd.tsv")                                                                                                                                                                                                           
s_df = pd.read_csv(tar_path, sep = '\t')
s_df['pmid'] = s_df['pmid'].astype(str)
s_df['geneId'] = s_df['geneId'].astype(str)
s_df['diseaseId'] = s_df['diseaseId'].astype(str)
```

```python
# read ann
in_f = '../data/ft_data/anns_n.txt'
anns = []
with open(in_f, 'r') as F:
    line = F.readline()
    while True:
        if len(line) == 0:
            break
        
        ann = line.strip().split("\t")
        if len(ann) > 5 and ';' in ann[5]:
            anns.append(ann[:])
            
        line = F.readline()
```

```python
len(anns)
```

```python
ann_df = pd.DataFrame(anns)
ann_df = ann_df[[0, 4, 5]]

ann_df['ori_id'] = ann_df.apply(lambda x: x[5].split(';')[0], axis=1)
ann_df[0] = ann_df[0].astype(str)
ann_df['ori_id'] = ann_df['ori_id'].astype(str)
ann_df = ann_df.drop_duplicates()
```

```python
ann_df
```

```python
merge_df = ann_df.merge(s_df, left_on=[0, 'ori_id'], right_on=['pmid', 'geneId'], how='left')
merge_df = merge_df[merge_df['2nd_label'] > 0]
```

```python
n_s_df = s_df.copy()
```

```python
for index, row in merge_df.iterrows():
    a, b, c = row['pmid'], row['geneId'], row['diseaseId']
    new_id = row[5].split(';')[1]
    
    print(a, b, c, row[5])
    print(new_id)
    
    n_s_df.loc[(n_s_df['pmid']==a) & (n_s_df['geneId']==b) & \
     (n_s_df['diseaseId']==c), 'geneId'] = new_id
```
